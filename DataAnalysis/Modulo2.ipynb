```python
# The first thing we need to do is import pandas library
import pandas as pd

# And we're going to need matplotlib 
import matplotlib.pylab as plt

# I'll need numpy for data replacement
import numpy as np
```

```python
# We'll need a dataset to work with it

filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
# It is desirable to have the dataset headers:

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
```

```python
# Let's examine the data

df.head(10)
```

```python
# The first thing we need to do is evaluate for missing data
# I'll replace every ? 
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
# I can mark all the null data

all_my_missing_data = df.isnull()
all_my_missing_data.head()
```

```python
# As indicated in the exercise, I can count how many rows are in each column:

for col in all_my_missing_data.columns.values.tolist():
  print(col)
  print(all_my_missing_data[col].value_counts())

  
```

```python
# Now, let's move towards how to handle data
# In previous lessons we seen how to handle with missed data
# This is two fold:
# Drop data or replace data
#   Drop data:
#   Drop the whole row
#   Drop the whole column
# Replace data
#   Replace it by the mean
#   Replace it by the frequency
#   Replace it based in other functions
# We need to evaluate carefully which techinque we'll use, for example, there are rows without price, these rows are useless for every prediction, thus, we can proceed to remove them from the dataset.

# Calculating the mean value for the normalized-losses column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of normalized-losses:", avg_norm_loss)

# Replace NaN with mean in normalized-losses
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)

# Calculate the mean value for the "bore" column
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)

#Replace "NaN" with the mean value in the "bore" column
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

###Question #1: 
Based on the example above, replace NaN in "stroke" column with the mean value.

```python
avg_stroke=df['stroke'].astype('float').mean(axis=0)
df["stroke"].replace(np.nan, avg_bore, inplace=True)
```

```python
# Calculate the mean for the hp column
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
#Replace NaN with the mean for peak-rpm
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
```

```python
# A nice built in function in pandas is idxmax() that allows us calculate the most common type:
df['num-of-doors'].value_counts().idxmax()
```

```python
# With this information, we can replace the NaN with the most frequent value
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# At the end it is necesary to drop all the rows that don't have data
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

Now, the next steps is checking and making sure that the whole data is in correct format

In this lesson, I learnd .dstype() and .astype() built in functions


```python
# To see all dastatypes within a dataframe
df.dtypes
```

```python
# What if I need to transform in another type of data?

df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")

# Let's check again how the old data types were converted to the new ones

df.dtypes
```

```python
# Let's standarize the data
# For example, maybe we need to transform mpg into L/100l,
df['city-L/100km'] = 235/df["city-mpg"]

# Now we can check it
df['city-L/100km'].head()
#It's already standarized
```

Question #2: 
According to the example above, transform mpg to L/100km in the column of "highway-mpg" and change the name of column to "highway-L/100km".

```python
df["highway-mpg"] = 235/df["highway-mpg"]
df.rename(columns={'highway-mpg':'highway-L/100km'}, inplace=True)
df["highway-L/100km"].head()
```

```python
# Maybe we need to normalize some data. This is to transforming existing data into a similar range.
# For instance we can update our features and normalized:
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

Question #3: 
According to the example above, normalize the column "height".

```python
df['height'] = df['height']/df['height'].max() 
df[["length","width","height"]].head()
```

```python
# Binning
# Binning is the process of transforming continous numerical variables into discrete categorical bins

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot


# For example:
df["horsepower"]=df["horsepower"].astype(int, copy=True)

# draw historgram of attribute "horsepower" with bins = 3
plt.pyplot.hist(df["horsepower"], bins = 3)

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")

# Now we have three bins (groups) for horsepower feature 
```

```python
# Let's review the dummy variable
# What is a dummy variable?
# It is a numerical variable used to label categories

dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
dummy_variable_1.rename(columns={'gas':'fuel-type-gas', 'diesel':'fuel-type-diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# Then, let's add this cols to the dataframe

df = pd.concat([df, dummy_variable_1], axis=1)

#And drop the original

df.drop("fuel-type", axis = 1, inplace=True)
```
