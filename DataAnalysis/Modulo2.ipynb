```python
# The first thing we need to do is import pandas library
import pandas as pd

# And we're going to need matplotlib 
import matplotlib.pylab as plt

# I'll need numpy for data replacement
import numpy as np
```

```python
# We'll need a dataset to work with it

filename = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
# It is desirable to have the dataset headers:

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(filename, names = headers)
```

```python
# Let's examine the data

df.head(10)
```

```python
# The first thing we need to do is evaluate for missing data
# I'll replace every ? 
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
# I can mark all the null data

all_my_missing_data = df.isnull()
all_my_missing_data.head()
```

```python
# As indicated in the exercise, I can count how many rows are in each column:

for col in all_my_missing_data.columns.values.tolist():
  print(col)
  print(all_my_missing_data[col].value_counts())

  
```
